```python
import os 
import pandas as pd
import json 

from collections import defaultdict

DOC_TYPE = 'PL'
DATAFRAME_RESULT_PATH = f'/home/administrator/Vasu/EXTRACTION_WORK/EXTRACTION_RESULTS/{DOC_TYPE}'
DOC_EXTRACTION_MODEL_RESULTS = f'/home/administrator/Vasu/EXTRACTION_WORK/EXTRACTION_RESULTS/{DOC_TYPE}/MODEL_RESULTS'
DOC_EXTRACTION_POST_PROCESSING_RESULTS = f'/home/administrator/Vasu/EXTRACTION_WORK/EXTRACTION_RESULTS/{DOC_TYPE}/POST_PROCESSING_RESULTS'
```

```python
df = pd.DataFrame(columns=['Document_Type','Image_Name','Field_Key','Model_BBOX','Model_Value','Post_Processing_BBOX','Post_Processing_Value'])
model_mapping_dict = defaultdict(list)
postprocessing_mapping_dict = defaultdict(list)
final_mapping_dict = defaultdict(list)
```

```python
for file in os.listdir(DOC_EXTRACTION_MODEL_RESULTS):
    file_name = file[:-18]
    file_path = os.path.join(DOC_EXTRACTION_MODEL_RESULTS,file)
    with open(file_path,'r') as f:
        try:
            data = json.load(f)             
            for key in data.keys():
                page_data = data[key]
                for page_wise_key in page_data.keys():
                    current_data_lst = page_data[page_wise_key]
                    for lst in current_data_lst:
                        model_mapping_dict[file_name].append([DOC_TYPE,page_wise_key,lst[1],lst[0]])
        except json.JSONDecodeError as e:
            print(f"Error decoding JSON from file {file_path}: {e}")
print(model_mapping_dict)
```

```python
for file in os.listdir(DOC_EXTRACTION_POST_PROCESSING_RESULTS):
    file_name = file[:-28]
    file_path = os.path.join(DOC_EXTRACTION_POST_PROCESSING_RESULTS,file)
    with open(file_path,'r') as f:
        try:
            data = json.load(f)
            for key in data['keys_extraction'].keys():
                if data['keys_extraction'][key] and 'value' in data['keys_extraction'][key].keys() and 'coordinate' in data['keys_extraction'][key].keys():
                    postprocessing_mapping_dict[file_name].append([DOC_TYPE,key,data['keys_extraction'][key]['coordinate'],data['keys_extraction'][key]['value']])
        except json.JSONDecodeError as e:
            print(f"Error decoding JSON from file {file_path}: {e}")
print(postprocessing_mapping_dict)
```

```python
for key in model_mapping_dict.keys():
    model_key_lst = model_mapping_dict[key]
    model_key_lst.sort(key=lambda x: (x[0], x[1]))
    
    combined_results = {}
    
    for item in model_key_lst:
        key_tuple = (item[0], item[1])
        if key_tuple in combined_results:
            combined_results[key_tuple][0].append(item[2])
            combined_results[key_tuple][1].append(item[3])
        else:
            combined_results[key_tuple] = [[item[2]], [item[3]]]
    
    combined_list = [[key_tuple[0], key_tuple[1], combined_results[key_tuple][0], combined_results[key_tuple][1]] for key_tuple in combined_results]
    model_mapping_dict[key] = combined_list

print(model_mapping_dict)
```

```python
for key in model_mapping_dict.keys():
    model_mapping_lst = model_mapping_dict[key]
    model_mapping_lst.sort(key=lambda x: (x[0], x[1]))
    postprocessing_lst = postprocessing_mapping_dict[key]
    postprocessing_lst.sort(key=lambda x: (x[0], x[1]))
    
    combined_results = []
    postprocessing_dict = {(item[0], item[1]): item for item in postprocessing_lst}

    for model_item in model_mapping_lst:
        key_tuple = (model_item[0], model_item[1])
        if key_tuple in postprocessing_dict:
            pp_item = postprocessing_dict[key_tuple]
            combined_results.append([
                model_item[0],
                model_item[1],
                model_item[2],
                pp_item[2],
                model_item[3],
                pp_item[3]
            ])
    
    final_mapping_dict[key] = combined_results
    
print(final_mapping_dict)
```

```python
rows = []
for image_name, fields in final_mapping_dict.items():
    for field in fields:
        document_type = field[0]
        field_key = field[1]
        model_bbox = field[2]
        postprocessing_bbox = field[3]
        model_value = field[4]
        postprocessing_value = field[5]
        
        rows.append({
            'Document_Type': document_type,
            'Image_Name': image_name,
            'Field_Key': field_key,
            'Model_BBox': model_bbox,
            'PostProcessing_BBox': postprocessing_bbox,
            'Model_Value': model_value,
            'PostProcessing_Value': postprocessing_value
        })

df = pd.DataFrame(rows)
csv_file_path = DATAFRAME_RESULT_PATH + f'/{DOC_TYPE}_BBOX&VALUES.csv'
df.to_csv(csv_file_path, index=False)
```
