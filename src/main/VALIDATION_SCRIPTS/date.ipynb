```python
import os 
import pandas as pd
import json 

from collections import defaultdict

DOC_TYPE = 'PL'
DATAFRAME_RESULT_PATH = f'/home/administrator/Vasu/EXTRACTION_WORK/EXTRACTION_RESULTS/{DOC_TYPE}'
DOC_EXTRACTION_MODEL_RESULTS = f'/home/administrator/Vasu/EXTRACTION_WORK/EXTRACTION_RESULTS/{DOC_TYPE}/MODEL_RESULTS'
DOC_EXTRACTION_POST_PROCESSING_RESULTS = f'/home/administrator/Vasu/EXTRACTION_WORK/EXTRACTION_RESULTS/{DOC_TYPE}/POST_PROCESSING_RESULTS'
```

```python
df = pd.DataFrame(columns=['Document_Type','Image_Name','Field_Key','Model_Date','Post_Processing_Date'])
model_mapping_dict = defaultdict(list)
postprocessing_mapping_dict = defaultdict(list)
final_mapping_dict = defaultdict(list)
```

```python
for file in os.listdir(DOC_EXTRACTION_MODEL_RESULTS):
    file_name = file[:-18]
    file_path = os.path.join(DOC_EXTRACTION_MODEL_RESULTS,file)
    with open(file_path,'r') as f:
        try:
            data = json.load(f)             
            for key in data.keys():
                page_data = data[key]
                for page_wise_key in page_data.keys():
                    if 'date' in page_wise_key:                        
                        if page_data[page_wise_key]:
                            potential_value_lst = page_data[page_wise_key]
                            for lst in potential_value_lst:                
                                postprocessing_mapping_dict[file_name].append([DOC_TYPE,page_wise_key,lst[0]])
        except json.JSONDecodeError as e:
            print(f"Error decoding JSON from file {file_path}: {e}")
```

```python
print(postprocessing_mapping_dict)
```

```python
for file in os.listdir(DOC_EXTRACTION_POST_PROCESSING_RESULTS):
    file_name = file[:-28]
    file_path = os.path.join(DOC_EXTRACTION_POST_PROCESSING_RESULTS,file)
    with open(file_path,'r') as f:
        try:
            data = json.load(f)
            for key in data['keys_extraction'].keys():
                if 'date' in key:
                    if data['keys_extraction'][key] and 'value' in data['keys_extraction'][key].keys():
                        model_mapping_dict[file_name].append([DOC_TYPE,key,data['keys_extraction'][key]['value']])
        except json.JSONDecodeError as e:
            print(f"Error decoding JSON from file {file_path}: {e}")
print(model_mapping_dict)
```

```python
print(model_mapping_dict)
```

```python
for key in postprocessing_mapping_dict.keys():
    key_mapping_lst = postprocessing_mapping_dict[key]
    key_mapping_lst.sort(key=lambda x: (x[0], x[1]))
    
    combined_results = {}
    
    for item in key_mapping_lst:
        key_tuple = (item[0], item[1])
        if key_tuple in combined_results:
            combined_results[key_tuple].append(item[2])
        else:
            combined_results[key_tuple] = [item[2]]
    
    result = []
    for key_tuple, values in combined_results.items():
        if len(values) > 1:
            combined_value = '~'.join(values)
            result.append([key_tuple[0], key_tuple[1], combined_value])
        else:
            result.append([key_tuple[0], key_tuple[1], values[0]])
    
    postprocessing_mapping_dict[key] = result

print(postprocessing_mapping_dict)
```

```python
for key in model_mapping_dict.keys():
    model_mapping_lst = model_mapping_dict[key]
    model_mapping_lst.sort(key=lambda x: (x[0], x[1]))
    postprocessing_lst = postprocessing_mapping_dict[key]
    postprocessing_lst.sort(key=lambda x: (x[0], x[1]))
    combined_result = []    
    postprocessing_dict = {(item[0], item[1]): item[2] for item in postprocessing_lst}
    for model_item in model_mapping_lst:
        model_key = (model_item[0], model_item[1])
        if model_key in postprocessing_dict:
            combined_value = [model_item[2], postprocessing_dict[model_key]]
        else:
            combined_value = [model_item[2]]
        combined_result.append([model_item[0], model_item[1]] + combined_value)
    
    final_mapping_dict[key] = combined_result

print(final_mapping_dict)
```

```python
rows = []
for image_name, fields in final_mapping_dict.items():
    for field in fields:
        document_type = field[0]
        field_key = field[1]
        post_processing_date = field[2]
        model_date = field[3] if len(field) > 3 else ''
        rows.append({
            'Document_Type': document_type,
            'Image_Name': image_name,
            'Field_Key': field_key,
            'Model_Date': model_date,
            'Post_Processing_Date': post_processing_date
        })
df = pd.concat([df, pd.DataFrame(rows)], ignore_index=True)
csv_path = DATAFRAME_RESULT_PATH + f'/{DOC_TYPE}_Date.csv'
df.to_csv(csv_path) 
```
